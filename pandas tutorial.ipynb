```python
class JpMorgan:
    def __init__(self, name):
        self.name = name

    def full_name(self, other):
        if isinstance(other, JpMorgan):
            return self.name + " " + other.name
        else:
            raise TypeError("Can only add JpMorgan objects")

obj1 = JpMorgan("rahul")
obj2 = JpMorgan("tiwari")

result = obj1.full_name(obj2)
print(result)
```
